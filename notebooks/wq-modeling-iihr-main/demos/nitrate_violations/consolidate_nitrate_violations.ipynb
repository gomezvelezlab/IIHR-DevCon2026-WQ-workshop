```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os 

dir_data = '/Users/jgomezvelez/Downloads/nitrate_violations'
```

```python
foldernames = [fn for fn in os.listdir(dir_data) if (not fn.endswith('.zip')) and (fn != '.DS_Store') and (not fn.endswith('.pkl')) and (not fn.endswith('.csv'))]
foldernames = sorted(foldernames)
foldernames
```

```python
data_list = []

for foldername in foldernames:
    year = int(foldername.split('Year')[1])
    print(year)

    # Get the list of files in the folder
    subfolders_level_1  = os.listdir(os.path.join(dir_data, foldername))
    subfolders_level_1 = [fn for fn in subfolders_level_1 if fn != '.DS_Store']
    subfolders_level_1 = sorted(subfolders_level_1)
    
    for subfolder_level_1 in subfolders_level_1:

        quarter = int(subfolder_level_1.split('Quarter')[1])

        subfolders_level_2 = os.listdir(os.path.join(dir_data, foldername, subfolder_level_1))
        subfolders_level_2 = [fn for fn in subfolders_level_2 if fn != '.DS_Store']
        subfolders_level_2 = sorted(subfolders_level_2)

        for subfolder_level_2 in subfolders_level_2:

            violation_type = int(subfolder_level_2.split('Violation (')[1].split(')')[0])

            subfolder_level_2_path = os.path.join(dir_data, foldername, subfolder_level_1, subfolder_level_2)

            contaminants = os.listdir(os.path.join(dir_data, foldername, subfolder_level_2_path))
            contaminants = [fn for fn in contaminants if (fn != '.DS_Store') and ('no results' not in fn)]
            contaminants = sorted(contaminants)

            for contaminant in contaminants:

                filename = os.listdir(os.path.join(subfolder_level_2_path, contaminant))
                filename = [fn for fn in filename if fn != '.DS_Store'][0]

                data = pd.read_excel(os.path.join(subfolder_level_2_path, contaminant, filename), skiprows=4)
                data['Year'] = year
                data['Quarter'] = quarter
                data['Violation'] = violation_type
                data['Contaminant'] = contaminant

                data_list.append(data)
    # break
```

```python
dataset = pd.concat(data_list)
dataset.head()
```

```python
dataset.to_pickle(os.path.join(dir_data, 'dataset_violations.pkl'))
dataset.to_csv(os.path.join(dir_data, 'dataset_violations.csv'))
```
